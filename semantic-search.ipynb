```python
count = 0  # we'll use the count to create unique IDs
batch_size = 50  # process everything in batches of 50

for i in tqdm(range(0, len(newdata), batch_size)):
    # set end position of batch
    i_end = min(i+batch_size, len(newdata))
    # get batch of lines and IDs
    abstracts_batch = [newdata[x]['Body'] for x in range(i,i+batch_size)]
    titles_batch = [newdata[x]['Title'] for x in range(i,i+batch_size)]
    ids_batch = [str(n) for n in range(i, i_end)]
    res = openai.Embedding.create(input=abstracts_batch, engine=MODEL)
    embeds = [record['embedding'] for record in res['data']]
    # prep metadata and upsert batch
    meta = [{'abstract': abstract, 'title': title} for abstract, title in zip(abstracts_batch, titles_batch)]
    meta
    to_upsert = zip(ids_batch,embeds, meta)
    index.upsert(vectors=list(to_upsert))
```

```python
query = "javascript error"

xq = openai.Embedding.create(input=query, engine=MODEL)['data'][0]['embedding']
```

```python
res = index.query([xq], top_k=5, include_metadata=True)
res
```
